```python
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

# For additional utilities
import numpy as np
import os
```

```python
source_folder = r"C:\Users\rohit\Desktop\hate\hateful_memes\img"
train_folder = r"C:\Users\rohit\Desktop\hate\hateful_memes\img\train"
validation_folder = r"C:\Users\rohit\Desktop\hate\hateful_memes\img\validation"

train_dir = os.path.join(source_folder, train_folder)
validation_dir = os.path.join(source_folder, validation_folder)

# Various Directory Paths
train_hateful_memes_dir = os.path.join(train_dir, 'hateful')
train_not_hateful_memes_dir = os.path.join(train_dir, 'not_hateful')
validation_hateful_memes_dir = os.path.join(validation_dir, 'hateful')
validation_not_hateful_memes_dir = os.path.join(validation_dir, 'not_hateful')
```

```python
#Checking Count of Images in Dataset

print('total training hateful images:', len(os.listdir(train_hateful_memes_dir)))
print('total training not hateful images:', len(os.listdir(train_not_hateful_memes_dir)))
print('total validation hateful images:', len(os.listdir(validation_hateful_memes_dir)))
print('total validation not hateful images:', len(os.listdir(validation_not_hateful_memes_dir)))
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255.,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

validation_datagen = ImageDataGenerator(rescale=1./255.)
```

```python
# Flow training images in batches of 10 using train_datagen generator
train_generator = train_datagen.flow_from_directory(
    train_dir,  # This is the source directory for training images
    classes=['hateful', 'not_hateful'],  # The classification
    target_size=(224, 224),  # All images will be resized to 224x224
    batch_size=10,
    class_mode='binary'  # Use binary labels
)

# Flow validation images in batches of 10 using valid_datagen generator
validation_generator = validation_datagen.flow_from_directory(
    validation_dir,  # This is the source directory for validating images
    classes=['hateful', 'not_hateful'],  # The classification
    target_size=(224, 224),  # All images will be resized to 224x224
    batch_size=10,
    class_mode='binary',  # Use binary labels
    shuffle=False
)
```

```python
from tensorflow.keras.applications import ResNet50

base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
```

```python
for layer in base_model.layers:
    layer.trainable = False
```

```python
from tensorflow.keras import layers

x = layers.Flatten()(base_model.output)

# Add a fully connected layer with 512 hidden units and ReLU activation
x = layers.Dense(512, activation='relu')(x)

# Add a dropout rate of 0.5
x = layers.Dropout(0.5)(x)

# Add a final sigmoid layer with 1 node for classification output
x = layers.Dense(1, activation='sigmoid')(x)

# Create the final model
model = Model(base_model.input, x)

# Print the model summary
model.summary()
```

```python
from tensorflow.keras.optimizers import RMSprop

model.compile(optimizer=RMSprop(lr=0.0001), loss='binary_crossentropy', metrics=['acc'])
```

```python
resnett = model.fit(
    train_generator,
    steps_per_epoch=480,  # Total number of steps (batches of samples) to draw from the generator per epoch
    epochs=10,
    verbose=1,
    validation_data=validation_generator,
    validation_steps=120  # Total number of steps (batches of samples) to draw from the generator for validation
)
```

```python
# Unfreeze some layers of the base model for fine-tuning
for layer in base_model.layers[-20:]:
    layer.trainable = True
```

```python
# Recompile the model with a lower learning rate
model.compile(optimizer=RMSprop(lr=1e-5), loss='binary_crossentropy', metrics=['acc'])
```

```python
# Continue training
resnett = model.fit(
    train_generator,
    steps_per_epoch=480,  # Total number of steps (batches of samples) to draw from the generator per epoch
    epochs=10,
    verbose=1,
    validation_data=validation_generator,
    validation_steps=120  # Total number of steps (batches of samples) to draw from the generator for validation
)
```

```python
val_loss, val_acc = model.evaluate(validation_generator)
print("Validation Loss:", val_loss)
print("Validation Accuracy:", val_acc)
```

```python
model.save('resnet50_hateful_memes_classifier.h5')
```

```python
import matplotlib.pyplot as plt

# Plot training & validation accuracy values
plt.plot(resnett.history['acc'])
plt.plot(resnett.history['val_acc'])
plt.title('Model accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(resnett.history['loss'])
plt.plot(resnett.history['val_loss'])
plt.title('Model loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```
